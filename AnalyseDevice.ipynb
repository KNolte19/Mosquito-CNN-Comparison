```python
import os

import pandas as pd
import numpy as np
import seaborn as sb

import scipy.stats as st
import sklearn.metrics as skm
import matplotlib.pyplot as plt

# Define costum color palette
colors = ["white", "firebrick", "gold", "darkseagreen", "seagreen"]
species_list = ["Ae. aegypti", "Ae.albopictus", "Ae.japonicus", "Ae.koreicus"]
custom_palette = sb.blend_palette(colors, as_cmap=True)
```

```python
def get_ci(performance_list):
    mean_performance = np.mean(performance_list)
    ci_performance = st.t.interval(0.95,
                                len(performance_list)-1,
                                loc=np.mean(performance_list),
                                scale=st.sem(performance_list))
    
    return round(mean_performance*100,2), (round(ci_performance[0]*100,2) , round(ci_performance[1]*100,2))
```

```python
df = pd.DataFrame(columns=["depiction", "trained_on", 'seed', 'pred_label', 'highest_score', 'sec_pred_label',
       'sec_highest_score', 'datasplit', 'true_label', 'correct', 'image_path'])

DIR = "results/device"
for model in os.listdir(DIR):
    if model.endswith(".pkl"):
        trained_device = model.split("_")[0]
        depiction = model.split("_")[1]
        seed = model.split("_")[2]
        
        df_temp = pd.read_pickle(os.path.join(DIR, model))
        
        df_temp["trained_on"] = [str(trained_device).capitalize()]*len(df_temp)
        df_temp["depiction"] = [str(depiction)]*len(df_temp)
        df_temp["seed"] = [int(seed)]*len(df_temp)
        
        df = pd.concat([df, df_temp])    
        
df["image_name"] = [path.split(os.sep)[-1].split(".")[0] for path in df["image_path"]]
df["device"] = [path.split("_")[0] for path in df["image_name"]]
df.drop(columns="image_path", inplace=True)
df = df.reset_index(drop=True)
```

```python
# Access the 'greens' colormap
cmap = plt.cm.get_cmap('Greens')
# Get the last color of the colormap
last_color = cmap(0.7)
```

```python
df_body = pd.read_pickle("results/demand/body_demand_experiment_perimg.pkl")
df_wing = pd.read_pickle("results/demand/wing_demand_experiment_perimg.pkl")

df_body["Depiction"] = "Body"
df_wing["Depiction"] = "Wing"

df_demand = pd.concat([df_body, df_wing])
df_demand["Training Method"] = df_demand["train_method"].str.replace("ft", "fine-tuning").replace("tf", "transfer-learning")
```

```python
fig, ax = plt.subplot_mosaic([['a', 'a', "a"], ['b', 'c', "d"]], sharey=True, layout='constrained', figsize=(10,6))

sb.lineplot(df_demand,x="N_per_species", y="accuracy", hue="Depiction",style="Training Method",
            palette=["steelblue", "darkseagreen"],markers=True,err_style="bars", ax=ax["a"])
sb.pointplot(data=df.loc[(df["depiction"] == "body") & (df["datasplit"] == "test")], x="trained_on", y="correct", errorbar="ci", linestyle="none",
             capsize=.1, linewidth=1.25, markersize=6, marker="D", ax= ax["b"], label="Body", color="steelblue", dodge=True)

sb.pointplot(data=df.loc[(df["depiction"] == "wing") & (df["datasplit"] == "test")], x="trained_on", y="correct", errorbar="ci", linestyle="none",
             capsize=.1, linewidth=1.25, markersize=6, marker="D", ax= ax["b"], label="Wing", color=last_color, dodge=True)

sb.pointplot(data=df.loc[(df["depiction"] == "body") & (df["datasplit"] == "test")], x="trained_on", y="correct", hue="device", errorbar="ci", linestyle="none",
             capsize=.1, linewidth=1.25, markersize=6, marker="D", ax= ax["c"], palette="Blues", markers=["D", "o", "d"], dodge=True)

sb.pointplot(data=df.loc[(df["depiction"] == "wing") & (df["datasplit"] == "test")], x="trained_on", y="correct", hue="device", errorbar="ci", linestyle="none",
             capsize=.1, linewidth=1.25, markersize=6, marker="D", ax= ax["d"], palette="Greens", markers=["D", "o"], dodge=True)

ax["a"].set_title("A")
ax["b"].set_title("B")
ax["c"].set_title("C")
ax["d"].set_title("D")

ax["b"].set_yticks(np.arange(0.3,1.01,.1))
ax["b"].set_ylim(0.3, .9)
ax["a"].grid(axis="y")
ax["b"].grid(axis="y")
ax["c"].grid(axis="y")
ax["d"].grid(axis="y")

ax["a"].set_ylabel("Accuracy [%]")
ax["b"].set_ylabel("Accuracy [%]")

ax["a"].set_xlabel("Images per Class")
ax["b"].set_xlabel("Type of Images trained on")
ax["c"].set_xlabel("Type of Images trained on")
ax["d"].set_xlabel("Type of Images trained on")

ax["a"].legend(ncols=2, fancybox=True, loc="lower right")
ax["b"].legend(title='Depiction', ncols=1, fancybox=True, loc="lower right")
ax["c"].legend(title='Device', ncols=2, fancybox=True)
ax["d"].legend(title='Device', ncols=1, fancybox=True)
plt.tight_layout()

plt.savefig("figures/boxplot_devicebytrained_comparison.svg", dpi=300)
```
